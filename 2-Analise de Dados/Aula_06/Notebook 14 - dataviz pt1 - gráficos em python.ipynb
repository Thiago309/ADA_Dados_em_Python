Vamos começar com uns gráficos simples...

A forma mais simples de criar gráficos com o matplotlib é a partir de **listas**,**arrays**, etc.

A ideia é que sejam dados os pontos **(x, y)** que são então plotados

```python
# pontos x do intervalo a ser plotado

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# função que faz o plot dos pontos
# argumentos: x, y
plt.plot(x, y, marker="o")
```

O matplotlib nos permite **alterar o tamanho do gráfico** e também **plotar mais de uma curva em um mesmo eixo**:

```python
# inicializando a figura com o tamanho alterado
# é importante definir isso como a variável "figura", pra que possamos
# salvar a imagem a seguir

plt.figure(figsize=(5, 3)) # altera o tamanho do grafico que será plotado. Seu padrão de tamanho é em Pixels.

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

Para fazermos gráficos em eixos separados, usamos o `plt.show()`, que finaliza a figura e a exibe

Também é possível **salvar a figura**:

```python
# salvando a figura

figure = plt.figure(figsize=(5, 3))

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

figure.savefig("./figura1.png")
```

Note que a imagem foi salva com umas bordas que podem ser indesejáveis... Pra corrigir isso, sete o parâmetro **bbox_inches = "tight"**

```python
# savefig com bbox_inches="tight" salva o grafico com bordas mais finas.

figure = plt.figure(figsize=(8, 4))

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

figure.savefig("./figura2.png", bbox_inches="tight")
```

Também é possível adicionar títulos ao gráfico e aos eixos

```python
figure = plt.figure(figsize=(8, 4))

plt.title("Função seno e cosseno", fontsize=14) # Nomeia o grafico.

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.xlabel("x")           # Legenda pro eixo X.
plt.ylabel("y = f(x)")    # Legenda pro eixo Y.

plt.show()

figure.savefig("./figura3.png", bbox_inches="tight")
```

É possível mover as legendas com os argumentos do **plt.legend()**

```python
figure = plt.figure(figsize=(8, 4))

plt.title("Função seno e cosseno", fontsize=14)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x), label="f(x) = sin(x)")  
plt.plot(x, np.cos(x), label="f(x) = cos(x)") 

plt.xlabel("x")         # Legenda pro eixo X.
plt.ylabel("y = f(x)")  # Legenda pro eixo Y.

plt.legend()            # plota a legenda do grafico.
plt.show()

figure.savefig("./figura4.png", bbox_inches="tight")
```

```python
figure = plt.figure(figsize=(8, 4))

plt.title("Função seno e cosseno", fontsize=14)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x), label="f(x) = sin(x)")
plt.plot(x, np.cos(x), label="f(x) = cos(x)")

plt.xlabel("x")
plt.ylabel("y = f(x)")

plt.legend(bbox_to_anchor=(1, 0.5), loc="center left") # indica que a caixa da descrição está 100% do eixo X e 50% do eixo Y.
plt.show()                                             # "center left" posiciona a caixinha de descrição para o lado esquedo do 
                                                       # grafico.
figure.savefig("./figura5.png", bbox_inches="tight")
```

Para maiores informações sobre legendas, [clique aqui!](https://matplotlib.org/tutorials/intermediate/legend_guide.html)

Você encontra outros exemplos de legenda [aqui.](http://jb-blog.readthedocs.io/en/latest/posts/0012-matplotlib-legend-outdide-plot.html)

É possível mudar a cor das curvas, bem como o estilo, usando os [argumentos de estilo](https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.lines.Line2D.html).

Dica: para escolha de cores, use o **color picker do google** para pegar o código HEX

```python
figure = plt.figure(figsize=(8, 4))

plt.title("Função seno e cosseno", fontsize=14)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x), label="f(x) = sin(x)", ls=":", lw=4, color="#97b3de")
plt.plot(x, np.cos(x), label="f(x) = cos(x)", ls="--", lw=2, color="#320b3b")

plt.xlabel("x")
plt.ylabel("y = f(x)")

plt.legend(bbox_to_anchor=(1, 0.5), loc="center left")
plt.show()

figure.savefig("./figura6.png", bbox_inches="tight")
```

# 3) Subplots com o Matplotlib

Às vezes não estamos interessados em exibir todos os dados em um mesmo gráfico. Porém, gostaríamos que os diferentes gráficos estivessem próximos para serem visualizados simultaneamente e compararados.

Com o matplotlib, é possível fazer diversos gráficos diferentes em blocos!

A principal forma de se fazer múltiplos gráficos na mesma figura é usando o **plt.subplots()**

Essa função tem como parâmetro um número com 3 dígitos: o primeiro diz quantas linhas teremos, o segundo diz quantas colunas, e o terceiro diz em qual gráfico estamos.

O contador de gráficos começa em 1, sendo que o primeiro gráfico é o do canto superior esquerdo. O índice cresce da esquerda para a direita, e de cima para baixo.

Para editar cada gráfico individualmente, basta chamar a plt.subplot passando o índice do gráfico desejado. Daquele ponto em diante, todas as operações serão sobre aquele gráfico. 

> Obs.: se você quiser deixar os plots mais espaçados, adicione o "plt.tight_layout()" no final!

Veja o exemplo abaixo, que plota 4 gráficos:

```python
plt.style.use('ggplot')
'''
2 - indica 2 linhas que cortam a grade do gráfico no eixo Y.
2 - indica 2 linhas que cortam a grade do gráfico no eixo X.
1 - indica 1 a ordem do grafico (primeiro gráfico a ser plotado).
'''

plt.subplot(221)
plt.plot(x, np.sin(x))
plt.title("Função seno")
plt.xlabel("x")
plt.ylabel("y = sin(x)")

plt.subplot(222)
plt.plot(x, np.cos(x))
plt.title("Função cosseno")
plt.xlabel("x")
plt.ylabel("y = cos(x)")

plt.subplot(223)
plt.plot(x, np.tanh(x), color="#801919")
plt.title("Função tanh")
plt.xlabel("x")
plt.ylabel("y = tanh(x)")

plt.subplot(224)
plt.plot(x, np.exp(x), color="#801919")
plt.title("Função exp")
plt.xlabel("x")
plt.ylabel("y = exp(x)")

plt.tight_layout()
```
